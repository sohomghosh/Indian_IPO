# If next day does not exist copy the current day information

```python
new_li = []
for i in range(nifty50_daily.shape[0]-1):
  current = nifty50_daily.iloc[i]['day']
  next = nifty50_daily.iloc[i+1]['day']
  days_diff = (next - current).days
  if days_diff>1:
    for dd in range(days_diff-1):
      new_li.append([nifty50_daily.iloc[i]['Price_nifty50_daily'],	nifty50_daily.iloc[i]['Open_nifty50_daily'],	nifty50_daily.iloc[i]['High_nifty50_daily'],	nifty50_daily.iloc[i]['Low_nifty50_daily'],	nifty50_daily.iloc[i]['Volume_nifty50_daily'],	nifty50_daily.iloc[i]['Chg%_nifty50_daily'],	current + datetime.timedelta(days=(1+dd))])
```

```python
new_weekend_holiday_df = pd.DataFrame(new_li, columns = nifty50_daily.columns)
new_weekend_holiday_df.shape
```

```python
nifty50_daily_with_weekends = nifty50_daily._append(new_weekend_holiday_df)
```
