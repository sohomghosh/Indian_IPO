```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
urls = []
texts = []
for i in range(2007,2025):
  url = 'https://www.chittorgarh.com/report/mainboard-ipo-list-in-india-bse-nse/83/?year=' + str(i)
  reqs = requests.get(url)
  soup = BeautifulSoup(reqs.text, 'html.parser')

  for link in soup.find_all('a'):
    link_ex = link.get('href')
    if 'IPO' in link.text and 'https://www.chittorgarh.com/ipo/' in link_ex:
      urls.append(link_ex)
      texts.append(link.text)
```

```python
df = pd.DataFrame({'url':urls, 'Issuer Company':texts}).drop_duplicates()
df
```

```python
all_ipos = pd.read_csv('all_ipos_2007_2024.csv')
all_ipos
```

```python
all_ipo_links = all_ipos.merge(df, on = 'Issuer Company')
del all_ipo_links['Compare']
chittogarh = pd.read_csv('chittorgarh_mainline_30Mar2024_structured.csv')
all_ipo_links
```
