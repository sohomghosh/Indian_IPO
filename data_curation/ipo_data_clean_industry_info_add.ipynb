```python
industry_info = pd.read_csv('India_PLC_Sector_Industry_2024_MAY.csv')
industry_info['Description'] = industry_info['Description'].apply(lambda x : str(x).lower().strip())
industry_info
```

```python
from difflib import get_close_matches
get_close_matches('appel', ['ape', 'apple', 'peach', 'puppy'], n = 1)
```

```python
company_industry_list = industry_info['Description'].to_list()
```

```python
matching_company = []
for company in final_data['Issuer Company']:
  company = company.replace(' IPO', '').lower().strip()
  try:
    match = get_close_matches(company, company_industry_list, n = 1)[0]
  except:
    match = ''
  matching_company.append(match)
```

```python
df = pd.DataFrame({'Issuer Company' : final_data['Issuer Company'], 'Description' : matching_company})
df_industry = df.merge(industry_info, on = 'Description')
df_industry
```
