```python
data_ipo_quarterwise_new = pd.read_excel('sucess_ipo_quarterwise.xlsx').rename(columns = {'quarter':'previous_quater', 'Success_Open' : 'previous_quarter_success_open', 'Success_Close' : 'previous_quarter_success_close', 'Success_High' : 'previous_quarter_success_high' })
data_ipo_quarterwise_new
```

```python
data_sort['previous_quater'] = data_sort['previous_quater'].apply(lambda x : str(x))
```

```python
final_df = data_sort.merge(data_ipo_quarterwise_new, on = 'previous_quater', how = 'left')
final_df = final_df.rename(columns = {'previous_quater' : 'previous_quarter'})
final_df
```

```python
final_df.to_excel('main_data_with_indics_nifty50daily_weekly_monthly_vix_daily_weekly_monthly_FITERED_CLEANED_industry_news_prevperformance_v6.xlsx', index = False)
```

```python
def dynamic_quater(close_date_now, days=90):
  final_df_use = final_df[(final_df['Close Date']<close_date_now) & (final_df['Close Date']>=close_date_now-pd.Timedelta(days, unit='D'))]
  li = final_df_use[['Success_Open', 'Success_Close', 'Success_High']].mean()
  return [li[0], li[1], li[2]]
```

```python
import numpy as np
```

```python
li_of_li = [[0,1,1]] # 2008 data added manually
for close_date in final_df["Close Date"][1:]: # 2008 data added manually so starting from 1st index (and not 0th index)
  li = dynamic_quater(close_date)
  if str(li[0]) == 'nan': # If there were no ipo inn previous quarter, we are considering the penultimate quarter
    li_of_li.append(li_of_li[-1])
  else:
    li_of_li.append(li)

li_of_li
```

```python
len(li_of_li)
```

```python
final_df['dynamic_last_90Day_success_open'] = [i[0] for i in li_of_li]
final_df['dynamic_last_90Day_success_close'] = [i[1] for i in li_of_li]
final_df['dynamic_last_90Day_success_high'] = [i[2] for i in li_of_li]
```
