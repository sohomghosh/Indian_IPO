```python
def list_to_str(x):
  x = str(x).replace(', nan', '')
  if isinstance(x, str) and x!='' and x!='nan' and x.startswith('[') and x.endswith(']') and x is not np.nan:
    return ' '.join(eval(x))
  else:
    return ''
```

```python
for col in  ['news_headline',	'news_synopsis',	'news_content',	'full_text_content']:
  main[col] = main[col].apply(lambda x : list_to_str(x)).astype(str)
```

```python
# Since less than 50% of instances have news, creating embeddings just for news won't make sense
main['full_text_context_with_news'] =  main[ 'full_text_content'].fillna('').astype(str) + " \n\n Related News: " + main['news_headline'].fillna('').astype(str) + "\n" + 	main['news_synopsis'].fillna('').astype(str) + "\n" +	main['news_content'].fillna('').astype(str)
```

```python
text_features = ['full_text_content', 'full_text_context_with_news', 'answer_of_question_1',	'answer_of_question_2',	'answer_of_question_3','answer_of_question_4','answer_of_question_5','answer_of_question_6','answer_of_question_7','answer_of_question_8','answer_of_question_9','answer_of_question_10','answer_of_question_11','answer_of_question_12','answer_of_question_13','answer_of_question_14','answer_of_question_15','answer_of_question_16','answer_of_question_17','answer_of_question_18','answer_of_question_19','answer_of_question_20','answer_of_question_21','answer_of_question_22','answer_of_question_23','answer_of_question_24','answer_of_question_25']
```

```python
for txt_feat in text_features:
  main[txt_feat] = 'classification: ' + main[txt_feat]
```

```python
for txt_feat in text_features:
  #embeddings = model.encode(main[txt_feat].to_list())
  batch_size = 8  # Adjust batch size as needed based on available memory
  embeddings = []
  for i in range(0, len(main), batch_size):
    batch = main[txt_feat][i : i + batch_size].to_list()
    batch_embeddings = model.encode(batch)
    embeddings.extend(batch_embeddings)
    gc.collect()
  pickle.dump(embeddings, open(f"/content/drive/MyDrive/datasets/Indian_IPO_Prediction/final_data/mainboard_embeddings/mainboard_{txt_feat}.pkl", "wb"))
```
